```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_state_qsphere
import numpy as np
%matplotlib inline
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1024
```

```python
qubits = QuantumRegister(2, name='qubits')
ancilla = QuantumRegister(2, name='ancilla')
classical = ClassicalRegister(2, name='classical')

qc = QuantumCircuit(qubits,ancilla,classical)
```

```python
#qc.h(ancilla)
qc.h(qubits[0])
qc.cx(qubits[0],qubits[1])
qc.barrier()
qc.draw('mpl')
```

```python
qc.count_ops()
```

```python
qc.depth()
```

```python
#qc.z(qubits[0])
qc.barrier()
qc.cx(qubits[0],ancilla[0])
qc.cx(qubits[1],ancilla[0])
qc.barrier()
qc.h(qubits)
#qc.h(ancilla[1])
qc.cx(qubits[0],ancilla[1])
qc.cx(qubits[1],ancilla[1])
#qc.h(ancilla[1])
qc.h(qubits)
qc.barrier()
qc.draw('mpl')
```

```python
#qc.h(ancilla)
qc.measure(ancilla,classical)
qc.draw('latex')
```

```python
qc.count_ops()
```

```python
qc.depth()
```

```python
qc = qc.reverse_bits()
```

```python
job = execute(qc, backend=backend, shots = shots)
results = job.result().get_counts()
plot_histogram(results)
```
